```python
def extract_frames(label, video_data, output_folder):

    yt = YouTube(video_data['URL'])
    stream = yt.streams.get_by_itag(video_data['itag'])

    # 비디오 다운로드
    video_file_name = f"{output_folder}/{label}/datanum_{video_data['data_num']}_video_{video_data['video_num']}.mp4"
#     video_file_name = f"{output_folder}/{label}/datanum_{video_data['data_num']}_video_{video_data['video_num']}.mp4"
#     stream.download(video_file_name)
    stream.download(f"{output_folder}/{label}", filename=f"datanum_{video_data['data_num']}_video_{video_data['video_num']}.mp4")

    cap = cv2.VideoCapture(video_file_name)
    start_frame = int((video_data['start_min']*60+video_data['start_sec'])*video_data['fps'])
    end_frame = int((video_data['finish_min']*60+video_data['finish_sec'])*video_data['fps'])

    # 프레임별 이미지 및 정보 저장을 위한 리스트
    frames_info = []

    # 시작 프레임까지 건너뜀
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    for frame_num in range(start_frame, end_frame + 1,4):
    # 프레임 읽기
        ret, frame = cap.read()
        if not ret:
            raise ValueError("프레임이 없음")

        # 이미지 파일로 저장
        frame_image_path =  f"{output_folder}/{label}/datanum_{video_data['data_num']}/video_{video_data['video_num']}_frame_{frame_num}.jpg"
        cv2.imwrite(frame_image_path, frame)

        # 프레임 정보 저장
        frames_info.append({
            'data_num' : video_data['data_num'],
            'video_num' : video_data['video_num'],
            'frame_inter' : 2,
            'frame_totalcount' : end_frame + 1 - start_frame,
            'frame_num': frame_num,
            'time': frame_num / video_data['fps'],
            'image_path': frame_image_path
        })

    # 동영상 캡처 해제
    cap.release()
    # 다운로드한 동영상 파일 삭제
    os.remove(video_file_name)
    
    return pd.DataFrame(frames_info)
```

```python
climb_data_frame_info = pd.DataFrame()

for index, row in video_metadata.iterrows():
    result_df = extract_frames('climb', row, "D:\Dataset")
    climb_data_frame_info = pd.concat([climb_data_frame_info, result_df], ignore_index=True)
climb_data_frame_info
```

```python
climb_data_frame_info.to_csv("./climb_data_frame_info.csv")
```
